```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [16,9]

ax = plt.gca()

df_soma_ano = pd.DataFrame(df_result.groupby('year_disembark')['num_slaves'].sum())
df_soma_ano.reset_index(level=0, inplace=True)

df_soma_ano.plot(kind='bar',x='year_disembark',y='num_slaves',ax=ax)
plt.xlabel('Ano de Desembarque')
plt.ylabel('Número de escravos')
ax.get_legend().remove()
plt.show()
```
